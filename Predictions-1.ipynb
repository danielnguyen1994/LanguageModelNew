### Creating the Prediction Script:

```python
# Importing the Libraries

from tensorflow.keras.models import load_model
import numpy as np
import pickle


import warnings
#warnings.filterwarnings('ignore')

import ipywidgets as widgets
from IPython.display import display, clear_output


import tensorflow as tf
#tf.get_logger().setLevel('ERROR')
#tf.autograph.set_verbosity(1)
# Load the model and tokenizer

model = load_model('nextword1.h5')
tokenizer = pickle.load(open('tokenizer1.pkl', 'rb'))

def Predict_Next_Words(model, tokenizer, text):
    """
        In this function we are using the tokenizer and models trained
        and we are creating the sequence of the text entered and then
        using our model to predict and return the the predicted word.
    
    """
    for i in range(4):
        sequence = tokenizer.texts_to_sequences([text])[0]
        
        sequence = np.array(sequence)
        preds2 = model.predict(sequence) 
        #print(preds2)
        
        pred3 = np.argsort(model.predict(sequence), axis=-1)[0]     
        #print(pred3)        
        
        
        preds = model.predict_classes(sequence) 
        
        resultsno = slider.value
        #print(preds)       
        predicted_word = ""
        predicted_word2 = ""
        
        for x in range(resultsno):
            
            predsort = pred3[len(pred3) - (x + 1)]
            
            for key, value in tokenizer.word_index.items():            
                if value == predsort:                
                    predicted_word2 = key 
                    print('Vorschlag ',(x + 1),': ', predicted_word2)
                    break
               
        return predicted_word2
```

```python
# button send

button_send = widgets.Button(
                description='Get Recommendation',
                tooltip='Send',
                style={'description_width': 'initial'}
            )

output = widgets.Output()

def on_button_clicked(event):
    with output:
        clear_output()        
        
        text = name.value
        try:       
            text = text.split(" ")
            text = text[-1]

            text = ''.join(text)
            Predict_Next_Words(model, tokenizer, text)
        except:
            print("Wort nicht im Wortschatz")
            
def on_change(change):
    with output:
        clear_output()
        
        try:       
            text = change.new.split(" ")
            text = text[-1]

            text = ''.join(text)
            Predict_Next_Words(model, tokenizer, text)
        except:
            print("Wort nicht im Wortschatz")
            
            
name.observe(on_change, names =["value"])
           
     

button_send.on_click(on_button_clicked)

vbox_result = widgets.VBox([ button_send, output])
```
